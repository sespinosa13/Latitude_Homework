```python
#Dependencies
import csv
import matplotlib.pyplot as plt
import requests
import pandas as pd
from config import api_key
import json
```

```python
cities_pd = pd.read_csv('worldcities.csv')
```

```python
random_cities = cities_pd.sample(n=500)
new_cities = random_cities.reset_index()
new_cities['City'].value_counts()
```

```python
new_cities['Temperature'] =''
new_cities['Wind'] = ''
new_cities['Humidity'] = ''
new_cities['Cloudiness'] = ''
new_cities.head()
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
```

```python
# Loop through the new_cities and run a temp/wind/humidity/cloudiness search for each city

row_count = 0

for index, row in new_cities.iterrows():
    query_url = f"{url}appid={api_key}&units={units}&q="

    city = row['City']
    
    print("Now retrieving city #%s: %s" %
          (row_count, new_cities.loc[index]["City"]))
    row_count += 1

    # make request, print url
    city_response = requests.get(query_url + city)
    print(city_response.url)
    # convert to json
    city_reponse = city_response.json()

    new_cities.set_value(index, "Temperature", city_response['main']['temp_max'])
    new_cities.set_value(index, "Wind", city_response['wind']['speed'])
    new_cities.set_value(index, "Humidity", city_response['main']['humidity'])
    new_cities.set_value(index, "Cloudiness", city_response['clouds']['all'])
```

```python
# Build a scatter plot for each data type
plt.scatter(new_cities['Latitude'],
            new_cities["Temperature"],
            edgecolor="black", linewidths=1, marker="o",
            alpha=0.8)

# Incorporate the other graph properties
plt.title("City Latitude vs. Max Temp")
plt.ylabel("Max Temperature (f)")
plt.xlabel("Latitude")
plt.grid(True)
#plt.xlim([-2.5, 202])
#plt.ylim([-2.5, 110000])

# Save the figure
plt.savefig("output_analysis/Temp_Latitude.png")

# Show plot
plt.show()
```
